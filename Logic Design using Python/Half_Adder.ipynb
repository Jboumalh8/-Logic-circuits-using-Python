```python
import gates
from prettytable import PrettyTable

S = 'XR(A,B)'
C = 'AND(A,B)'

table = PrettyTable()
table.field_names = ["A", "B", "C", "S"]

print("Below is the truth table for a Half-Adder:")



for A in range(2):
    for B in range(2):
        S1 = eval(gates.sub(S))
        C1 = eval(gates.sub(C))
        table.add_row([A,B,C1,S1])

print(table)
```
