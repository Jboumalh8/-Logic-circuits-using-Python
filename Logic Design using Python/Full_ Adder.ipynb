```python
import gates
from prettytable import PrettyTable

S = 'OR(AND(NOT(A), XR(B, Ci)), AND(A, NOT(XR(B,Ci))))'
C = 'OR(AND(B,Ci), AND(A, XR(B,Ci)))'
table = PrettyTable()
table.field_names = ["A", "B", "Ci", "Co", "S"]

print("Below is the truth table for full adder")

print(gates.sub(S))

for A in range(2):
    for B in range(2):
        for Ci in range(2):
            S1 = eval(gates.sub(S))
            C1 = eval(gates.sub(C))
            table.add_row([A,B,Ci,C1,S1])
print(table)
```
