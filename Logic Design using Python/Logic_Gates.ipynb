# OR gate

```python
def OR(A,B):
    
    if A == 0 and B == 0:
        return 0
    else:
        return 1
```

```python
print("Enter Boolean 1 or 0")
A = int(input("A = "))
B = int(input("B = "))

print("Q = " ,OR(A,B))
```

```python
from prettytable import PrettyTable
table = PrettyTable(["A", "B", "Q"])

for A in range(2):
   
    for B in range(2):
        Q = OR(A,B)
        table.add_row([A,B,Q])
        
print("OR Gate")

print(table)
```

NOT GATE
